```python
from datetime import datetime
import re
import logging
import pprint
import collections
b = datetime.strptime('05:00:00', "%H:%M:%S").time()
a = datetime.strptime('00:00:00', "%H:%M:%S").time()
data = collections.OrderedDict()
sus_data ={}
irr_data = {}
gl_data = {}
dt_data = {}
with open ("userlog.log", "r") as f:
   for line in f:
        x = line.split()
        time = datetime.strptime(x[1], "%H:%M:%S").time()
        if x[4] in data:
            if x[0] in data [x[4]]:
                data[x[4]][x[0]][time] = (x[2],x[3])
            else: 
                data[x[4]][x[0]] = {}
                data[x[4]][x[0]][time] = (x[2],x[3])
        else:
            data[x[4]] = {}
            data[x[4]][x[0]] = {}
            data[x[4]][x[0]][time] = (x[2],x[3])
pprint.pprint(data)
#Loop through all dictionaries for all steps
for user in data:
    sus_data[user] = 0
    irr_data[user] = 0
    gl_data[user] = 0
    #Domain Logic
    domain = user.split('@')[1]
    if domain in dt_data:
        dt_data[domain] += 1
    else:
        dt_data[domain] = 1
    
    
    for date in data[user]:
        num_logins = 0
        num_logouts = 0
        early_logins = False
    
        for time in data[user][date]:
            if data[user][date][time][0]== "login":
                num_logins +=1
                if a < time < b:
                    early_logins = True
            if data[user][date][time][0]== "logout":
                num_logouts +=1
        # Check if they logged in too many times or if they logged in early
        if num_logins >= 5 or early_logins == True:
            sus_data[user] +=1
        #Check if they logged in more than they logged out
        if num_logouts < num_logins:
            irr_data[user] +=1
        #Check if they logged out more than they logged in
        if num_logouts > num_logins:
            gl_data[user] +=1
#Suspicious Activity
with open ("sustxt.txt", "w") as f:
    f.write("Suspicious Activity\n\n")
    for user in data:
        f.write (str(user) + "\n")
        f.write("No. of Suspicious Days " + str(sus_data[user]) +"\n")
        for date in data[user]:
            f.write('\t'+ str(date) + "\n")
            for time in data[user][date]:
                f.write('\t\t'+ str(time) + "\t"+ str(data[user][date][time][0]) + "\t" + str(data[user][date][time][1])+"\n" )
#Irresponsible Activity
with open ("irrtxt.txt", "w") as f:
    f.write("Irresponsible Activity\n\n")
    for user in data:
        f.write (str(user) + "\n")
        f.write("No. of Irresponsible Days " + str(irr_data[user]) +"\n")
        for date in data[user]:
            f.write('\t'+ str(date) + "\n")
            for time in data[user][date]:
                f.write('\t\t'+ str(time) + "\t"+ str(data[user][date][time][0]) + "\t" + str(data[user][date][time][1])+"\n" )
#Glitches Activity
with open ("gltxt.txt", "w") as f:
    f.write("Glitches \n\n")
    for user in data:
        f.write (str(user) + "\n")
        f.write("No. of Glitches " + str(gl_data[user]) +"\n")
        for date in data[user]:
            f.write('\t'+ str(date) + "\n")
            for time in data[user][date]:
                f.write('\t\t'+ str(time) + "\t"+ str(data[user][date][time][0]) + "\t" + str(data[user][date][time][1])+"\n" )
#Domain Activity
with open("dttxt.txt", "w") as f:
    f.write("Domain Count \n\n")
    for domain in dt_data:
        f.write(str(domain) +"\t" + str(dt_data[domain]) + "\n")
        

```
